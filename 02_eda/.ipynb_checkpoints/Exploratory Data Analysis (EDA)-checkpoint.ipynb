```python
#check up details of the sign_up date
print("data_type:",df_customers["signup_date"].dtype)
print(df_customers["signup_date"].describe())

print("min :",df_customers["signup_date"].min())
```

```python
# Analyze monthly customer signups to track user growth over time

# 1. Convert date column to datetime
df_customers['signup_date'] = pd.to_datetime(df_customers['signup_date'])

df_customers['signup_date'].dtype
```

```python
# Group the customer data by signup month, then count how many customers signed up in each month.
monthly_signup = df_customers.groupby(df_customers['signup_date'].dt.to_period("M")).size()
```

```python
#Sort  the monthly_signup from Highest to Lowest
monthly_signup.sort_values(ascending = False)
```

```python
# Plot monthly customer signup_date trend
import matplotlib.pyplot as plt

monthly_signup.plot(kind='line', title='Monthly Customer Signups', figsize=(12, 6))

# Plot with labels
monthly_signup.plot(kind='line', figsize=(12, 6), title='Monthly Customer Signups')

# Label the Y-axis
plt.ylabel("Number of New Customers")

# Label the X-axis (optional for polish)
plt.xlabel("Signup Month")
plt.grid()

# Show the plot
plt.show()
```

### Comparing Signups by Region over Time

```python

# Ensure signup_date is datetime
#df_customers['signup_date'] = pd.to_datetime(df_customers['signup_date'])

# Create signup_month column again with no "size()"
df_customers['signup_month'] = df_customers['signup_date'].dt.to_period("M")

#  Group by month and region
signup_region = df_customers.groupby(['signup_month', 'region']).size().unstack(fill_value=0)

# Plot
signup_region.plot(figsize=(14, 7), title="Monthly Customer Signups by Region")
plt.xlabel("Signup Month")
plt.ylabel("Number of Signups")
plt.legend(title="Region", bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid()
plt.tight_layout()
plt.show()
```

What to look for:

Did Lagos spike in any month?

Did a new region suddenly drop off?

Could be due to marketing effort, CRM targeting, or service outage.

## Compare Signups by Industry over Time

### Monthly Customer Acquisition(sign_ups) by Industry(Sector)

```python
import matplotlib.pyplot as plt

# Convert signup_date to datetime
df_customers['signup_date'] = pd.to_datetime(df_customers['signup_date'])

# Create a new 'signup_month' column
df_customers['signup_month'] = df_customers['signup_date'].dt.to_period("M")

# Group by both month and industry
signup_trend = df_customers.groupby(['signup_month', 'industry']).size().unstack(fill_value=0)

# Plot all industries on a single line plot
signup_trend.plot(figsize=(14, 7), title=" Monthly Signups by Industry")

# Label axes
plt.xlabel("Signup Month")
plt.ylabel("Number of New Customers")
plt.legend(title="Industry", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid()
# Show plot
plt.show()
```

What to investigate:

Any dominant industry like "Logistics" or "Retail"?

Sudden spikes in one industry = maybe targeted CRM campaign.

Drops may point to churn-sensitive sectors.

###  To Identify Spikes or Drops: flag months with spikes (e.g., above average)

```python
# Total signups per month:how many customers signed up each month.
monthly_signup = df_customers.groupby('signup_month').size()

# Calculate mean + std dev for anomaly thresholds
mean = monthly_signup.mean()
std = monthly_signup.std()

# Flag potential spikes or drops:average signups per month and how much variation (spread) exists.
# We will set a statistical baseline to detect anomalies.

#  Spike = unusually high signups (more than 1 standard deviation above the mean)
# Drop = unusually low signups (more than 1 std below)
# This helps flag: 
# Unexpected surges â†’ maybe a viral campaign?
# Weird drops â†’ maybe an app bug or downtime?

spikes = monthly_signup[monthly_signup > mean + std]
drops = monthly_signup[monthly_signup < mean - std]

print("ðŸš€ Spike Months:\n", spikes)
print("ðŸ“‰ Drop Months:\n", drops)
```

```python
monthly_signup.plot(figsize=(12, 6), title="ðŸ“ˆ Monthly Total Signups")
plt.axhline(mean, color='gray', linestyle='--', label='Average')
plt.scatter(spikes.index.to_timestamp(), spikes.values, color='green', label='Spike', zorder=5)
plt.scatter(drops.index.to_timestamp(), drops.values, color='red', label='Drop', zorder=5)
plt.xlabel("Month")
plt.ylabel("Number of Signups")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

Then ask: Did a CRM campaign or product launch happen around those spike months?

# 3. Top Merchants by Transaction Volume

####  Group transaction data by merchant and aggregate:
####  'count' to see number of transactions
####  'sum' to see total transaction value

```python
#  Group transaction data by merchant and aggregate:
# - 'count' to see number of transactions
# - 'sum' to see total transaction value

# Load transaction data
df_transactions = pd.read_csv("../Data Sources/data/transactions.csv")
merchant_stats = df_transactions.groupby('merchant_id')['amount'].agg(['count', 'sum'])

#  Sort merchants by total transaction amount (descending) and select top 10
merchant_stats = merchant_stats.sort_values('sum', ascending=False).head(10)
merchant_stats
```

```python
# Horizontal bar chart of top 10 merchants by total transaction volume
# Reset index for plotting
merchant_stats_reset = merchant_stats.reset_index().sort_values('sum', ascending=True)

# Lollipop chart
plt.figure(figsize=(12, 6))
plt.hlines(y=merchant_stats_reset['merchant_id'], xmin=0, xmax=merchant_stats_reset['sum'], color='skyblue')
plt.plot(merchant_stats_reset['sum'], merchant_stats_reset['merchant_id'], "o")

plt.title('Top 10 Merchants by Transaction Volume (Lollipop Style)', fontsize=14)
plt.xlabel('Total Transaction Value')
plt.ylabel('Merchant ID')
plt.tight_layout()
plt.grid(axis='x')
plt.show()
```
